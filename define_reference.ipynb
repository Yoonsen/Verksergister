```python
import dhlab.module_update as mu
import dhlab.api.dhlab_api as api
import dhlab.text as dh
import pandas as pd
import verk as v
```

```python
mu.css()
```

Sample a corpus from each writing period

```python
api.get_reference()
```

```python
corpus = {
    year : api.get_reference(corpus="digibok", from_year = year, to_year = year_to) for year, year_to in [(1800, 1840), (1850, 1900), (1920,1940), (1990,2000)]}
```

Collect top words

```python
topwords = {year:dh.Counts(corpus = corpus[year]).counts.sum(axis=1).sort_values(ascending=False).head(1500) for year in corpus}
```

Turn it into a dataframe

```python
res = pd.DataFrame({year:corpus[year]["freq"] for year in corpus})
```

```python
for x in res:
    res[x] = res[x]/res[x].sum()
```

```python
averages = pd.DataFrame()
```

```python
res = res*100
```

```python
m = res.mean(axis = 1)
```

```python
s = res.std(axis = 1)
```

```python
check = lambda x: m.loc[x] - s.loc[x]
```

```python
check('af')
```

```python
words = [x for x in res.index if check(x) < 0.1 and x.isalpha()]
```
